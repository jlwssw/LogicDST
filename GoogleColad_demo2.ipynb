## Setup Section
This section includes funnctions we will use in this notebook. We need to "activate" these functions, **ONLY ONCE**, before working on the tasks after this section, by hovering our mouse below and then clicing the **run cell** button.

````python
#The represent the function view0f as a set of pair
# Python does not directly support definition of function in this way
# so we need do addition work to allow a function call.or use.it as a function
# in the prefix form. Code written around 2023. This comment Feb 2026
viewsOf = {("Never Be the Same", 3), ("Havana", 20), ("Love You Like A Love Song", 7),("Come & Get It", 8)}
singerOf = {("Never Be the Same", "Camila Cabello"), ("Havana", "Camila Cabello"), ("Love You Like A Love Song", "Selena Gomez"),("Come & Get It", "Selena Gomez")}
# Derived from the two data sets above (same data, summarized by singer)
viewsOfSinger = {("Camila Cabello", 23), ("Selena Gomez", 15)}
class MyFunctionAsSet:
    def __init__(self, givenSet):
        self.data = givenSet
    def __call__(self, input):
        image = {y for (x, y) in self.data if x == input}
        if len(image) == 1: output = next(iter(image))
        else:
          output = None
          print("Sorry, I don't have the information for \"{input}\"") # None
        return output
viewsOf = MyFunctionAsSet(viewsOf)
singerOf = MyFunctionAsSet(singerOf)
viewsOfSinger = MyFunctionAsSet(viewsOfSinger)
print("Functions are activated.")
````

## Errors in expressions
As human being, we make mistakes often. So do we when typing the expressions. Here we give two types of errors.

### Typos in experssions 1 - syntax errors
One example is typo in the function name (see expression in the code cell below). In this case, google colab will report an error. This kind of errors are called **syntax errors**. The typo is often called a **bug** of the expression.

Once we recognize an error occuring, there are two things we can do.
1. We should be able to read the error message and use it to correct our expression.
2. We can click the button `Explain error`. A window will pop up from the bottom middle of google colab. It is called the **Colab AI** or **Gemini in Colab** or **Gemini pane**. The window is normally small. Hover your mouse on the text and scroll down (and up) to say all the text. Many times, it explains the error and propose solutions. It is a good idea for you to manually correct your expression although you can simply let google colab to the change by clicking the "Accept" option.  To close the window, click the **icon** below the window, which is often called **sparkle icon**.

The process of locating and correcting the bug(s) of an expression is called **debugging**.
